```python
import pandas as pd
import numpy as np
file= 'https://docs.google.com/spreadsheets/d/1jQPEsk2YCZwJjdjQvP_PpQQxYP5MofJ5/export'
df= pd.read_excel(file)
df.head()
```

```python
# knowing the stats
print(df.shape)
df.describe()
```

```python
#checking for missing values
df.isna().sum()
```

```python
# Heatmap to understand correlations among variables
import matplotlib.pyplot as plt
import seaborn as sns
cor= df.corr()
plt.figure(figsize=(9,7))
sns.heatmap(cor,annot=True)
plt.show()
```

```python
# Boxplot to determine outliers and stats
from sklearn.preprocessing import normalize 
df_norm= pd.DataFrame(normalize(df),columns=df.columns)
plt.figure(figsize=(12,7))
sns.boxplot(data=df_norm)
plt.xticks(rotation=90)
```

```python
# Distribution plot to analyze values in different columns
a=df_norm.columns
for i in a:
    sns.distplot(df_norm[i],label=i,kde=True,color='g')
    plt.legend(loc='upper right')
    plt.show()
```

```python
# pair plot to check every possible relation between 2 variables
sns.pairplot(data=df)
```

```python
#checking majority quality of wine
sns.countplot(x="quality",data=df)
```

```python
# Acidity and alcohol content accord to wine quality, does low acidity makes a good wine?
plt.figure(figsize=(10,7))
sns.scatterplot(x="volatile acidity",y="alcohol",hue="quality",data=df)
plt.show()
```
